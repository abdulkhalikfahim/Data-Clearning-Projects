```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#read csv file
a=pd.read_csv("ndvi_1.csv")

#convert csv into dataframe
#caution: DataFrame method is case sensitive
df=pd.DataFrame(a)

#rename default column
df=df.rename(columns={"system:time_start":"Time"})

#convert time column to datetime format

df['Time']=pd.to_datetime(df['Time'])
df.index=df['Time']
del df['Time']


rolling_mean=df.rolling(7).mean()
rolling_std=df.rolling(8).mean()

#sns.lineplot(df)

plt.plot(df,color='indigo')
plt.plot(rolling_mean,color="red")
plt.plot(rolling_std,color="black")

#plot file


plt.savefig("new",dpi=1200)
```

```python
plt.plot(rolling_mean.iloc[6:])
plt.xlabel('Time')
plt.ylabel('NDVI Moving Average')
plt.title('Moving Average NDVI,freq=5')
plt.savefig('Moving average',dpi=1200)
```
